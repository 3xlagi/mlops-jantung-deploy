impor modul dan library

```python
import os
import sys
from absl import logging
from tfx.orchestration.beam.beam_dag_runner import BeamDagRunner

from modules.components import init_components
from modules.pipeline import init_local_pipeline
```

Mendefinisikan Path Direktori

```python
PIPELINE_NAME = "rahmatdi-pipeline"

# Direktori input data dan modul
DATA_ROOT = "data"
TRANSFORM_MODULE_FILE = "modules/transform.py"
TRAINER_MODULE_FILE = "modules/trainer.py"

# Direktori output
OUTPUT_BASE = "output"
serving_model_dir = os.path.join(OUTPUT_BASE, 'serving_model')
pipeline_root = os.path.join(OUTPUT_BASE, PIPELINE_NAME)
```

Pipeline Orchestrator

```python
logging.set_verbosity(logging.INFO)

# Inisiasi seluruh komponen yang sudah kita buat di modules/components.py
components = init_components(
    data_dir=DATA_ROOT,
    transform_module=TRANSFORM_MODULE_FILE,
    training_module=TRAINER_MODULE_FILE,
    training_steps=5000,
    eval_steps=1000,
    serving_model_dir=serving_model_dir,
)

# Inisiasi dan jalankan pipeline menggunakan Apache Beam
pipeline = init_local_pipeline(components, pipeline_root)
BeamDagRunner().run(pipeline=pipeline)
```
