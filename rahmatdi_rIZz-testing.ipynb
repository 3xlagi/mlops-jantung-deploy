```python
import requests
import base64
import tensorflow as tf

URL = "https://mlops-jantung-deploy-production.up.railway.app/v1/models/rahmatdi-model:predict"

def _bytes_feature(value):
    """Mengembalikan bytes_list dari string / byte."""
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value.encode('utf-8')]))

def _float_feature(value):
    """Mengembalikan float_list dari float / double."""
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    """Mengembalikan int64_list dari bool / enum / int / uint."""
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

# ambil 1 sampel data dummy dari dataset heart.csv (baris pertama)
example = tf.train.Example(features=tf.train.Features(feature={
    'Age': _int64_feature(40),
    'Sex': _bytes_feature('M'),
    'ChestPainType': _bytes_feature('ATA'),
    'RestingBP': _int64_feature(140),
    'Cholesterol': _int64_feature(289),
    'FastingBS': _int64_feature(0),
    'RestingECG': _bytes_feature('Normal'),
    'MaxHR': _int64_feature(172),
    'ExerciseAngina': _bytes_feature('N'),
    'Oldpeak': _float_feature(0.0),
    'ST_Slope': _bytes_feature('Up')
}))

# Serialisasi ke dalam string dan encode ke Base64
serialized_example = example.SerializeToString()
b64_example = base64.b64encode(serialized_example).decode('utf-8')

payload = {
    "signature_name": "serving_default",
    "instances": [
        {
            "examples": {"b64": b64_example}
        }
    ]
}

# Kirim request POST ke endpoint
response = requests.post(URL, json=payload)

print("Status Code:", response.status_code)
print("Hasil Prediksi:", response.json())
```
